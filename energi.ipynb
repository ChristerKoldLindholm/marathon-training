```python
energy_data = pd.read_csv('energi.csv')
```

```python
df = energy_data.copy()
df.rename(columns={'energi (kJ/100g)':'energi'
    , 'energi/portion':'portion', 'protein/100g':'protein'}, inplace=True)
df.sort_values(by='portion', ascending=False, inplace=True)

df
```

```python
metab = 10_500

def e(food: str):
    return df.loc[df['fødevare'] == food, 'portion'].values[0]
```

```python
dinner_meat = sum([e('rugbrød')*2, e('makrel')*3, e('mayonnaise')
    , e('appelsin'), e('banan'), e('gulerødder')*4
    , e('kyllingbryst'), e('agurk')*2, e('majs')*2
    , e('cremefraiche'), e('tun')/3
    , e('kaffe m. mælk')*6, 
    ])

normal_day = sum([e('rugbrød')*2, e('æg'), e('avocado'), e('mayonnaise') * 2
    , e('banan'), e('gulerødder')*3, e('tun')/3, e('cremefraiche')
    , e('agurk'), e('majs'), e('kikærter'), e('kaffe m. mælk')*5
    , e('vindruer') 
    ])

home_day = sum([e('rugbrød')*2, e('æg'), e('avocado'), e('mayonnaise') * 2
    , e('banan'), e('gulerødder')*4, e('tun')/3, e('cremefraiche')
    , e('agurk'), e('majs')*2, e('kikærter'), e('kaffe m. mælk')*5
    , e('vindruer') 
    ])

work = sum([e('kaffe m. mælk')*3, e('rugbrød'), e('flute'), e('mayonnaise')
    , e('roastbeef'), e('svinekød'), e('cremefraiche'), e('kartofler'), e('rugbrød')*2, e('avocado')
    , e('æg'), e('rejer'), e('mayonnaise')*2, e('Cocio')*3, e('vindruer'), e('majs')
    ])
```

```python
work_fasting = sum([e('kaffe m. mælk')*4, e('mandler'), e('banan')*2, e('agurk')*2, e('kartofler'),
    e('kalv'), e('gulerødder')*4])

work_fasting_small = sum([e('kaffe m. mælk')*4, e('mandler'), e('banan')*3, e('gulerødder')*3,
    e('rugbrød')*2, e('avocado'), e('mayonnaise'), e('rejer'), e('karrysild'), e('blåbær')])
```

```python
diet_list = [dinner_meat, normal_day, home_day, work, work_fasting, work_fasting_small]
diet_names = ['dinner_meat', 'normal_day', 'home_day', 'work', 'work_fasting', 'work_fasting_small']
metab_covered = [round(diet/metab,2) for diet in diet_list]
diet_table = pd.DataFrame({'diet': diet_names, 'energy': diet_list, 'metab_covered': metab_covered})

diet_table
```
